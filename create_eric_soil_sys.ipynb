```python
import pandas as pd
import os
%config InlineBackend.figure_format = 'retina'
```

```python
# Load the uploaded file
file_path = 'eric_soil_file.csv'
df_data = pd.read_csv(file_path)

df_data = df_data.T.reset_index().T 
df_data

# Create directories for eric_soil_sys_1 to eric_soil_sys_5 and save the formatted data
file_paths = []
for i in range(1, 6):
    os.makedirs(f'eric_soil_sys_{i}', exist_ok=True)
    formatted_data_txt_file_path = f'eric_soil_sys_{i}/eric_soil_sys_{i}.txt'
    df_data.to_csv(formatted_data_txt_file_path, sep=' ', index=False, header=False)
    file_paths.append(formatted_data_txt_file_path)

# # Output the file paths
file_paths
```
